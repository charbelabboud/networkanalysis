```python
from neo4j.v1 import GraphDatabase
import pandas as pd
```

```python
driver = GraphDatabase.driver("bolt://127.0.0.1:7687",auth=basic_auth("neo4j", "boula1994"))
session = driver.session()

df = pd.read_csv("A.csv")
df.head()
```

```python
df.iloc[3][0]
```

```python
x = len(df.index)

# def add_person(self, name):
#     with self._driver.session() as session:
#         session.run("CREATE (a:Person {name: $name})", name=name)

for c in df["Unnamed: 0"]:
     session.run("create (:" + c.replace(" ","").replace("-","").replace("&","") + ")")

for i in range(x):
    for j in range(x):
          if(df.iloc[i][j+1]  != 0 ):
                session.run("MATCH (a),(b) WHERE a = '" + df.iloc[i][0] + "' AND b = '" + df.iloc[j][0] 
                            + "' CREATE (a)-[r:RELTYPE]->(b) RETURN r")
```

```python
result = session.run("MATCH (a),(b) WHERE a = '" + df.iloc[i][0] + "' AND b = '" + df.iloc[j][0] 
                            + "' CREATE (a)-[r:RELTYPE]->(b) RETURN r")
```
